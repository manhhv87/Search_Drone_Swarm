```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
size_fig = 10
df = pd.read_csv('result.csv')
df.columns = ['Execucao', 'Nº Drones', 'Nº Obstáculos', 'algoritmo', 'tempo_alvo', 'tempo_missao']
df = df.drop(columns=['Execucao'])
df['tempo_missao'] = df['tempo_missao'].replace('Goal not reached',np.NaN)
df['tempo_missao'] = df['tempo_missao'].astype('float64')
df['algoritmo'] = df['algoritmo'].replace('DefineTargetScan','Alvo Conhecido')
df['algoritmo'] = df['algoritmo'].replace('RowScan','Alvo Desconhecido')
df.head()
```

```python
#plt.figure(figsize=(20, 16))
sns.set_theme(style="whitegrid", palette="muted")
h = sns.jointplot(
    data=df.loc[(df['Nº Obstáculos'] == 10) & (df['algoritmo'] == 'Alvo Conhecido')],
    x="tempo_alvo", y="tempo_missao", hue="Nº Drones",
    kind="kde", height=size_fig
)
h.set_axis_labels('Tempo de Localização do Alvo (seg)', 'Tempo de Missão (seg)', fontsize=16)
h.ax_joint.set_xlabel('Métricas de Alvo Conhecido com 10 obstáculos', fontweight='bold')
```
